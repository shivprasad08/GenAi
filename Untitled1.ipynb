```python
import pretty_midi
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Load and Parse the MIDI File ---
# Load the MIDI file using the pretty_midi library [cite: 17, 36]
try:
    midi_data = pretty_midi.PrettyMIDI('/content/ACDC.Thunderstruck K.mid')
    print("MIDI file 'example.mid' loaded successfully!")
except IOError:
    print("Error: 'example.mid' not found. Please add a MIDI file to your directory.")
    # In case of error, create a dummy object to avoid further crashes
    midi_data = None

if midi_data:
    # --- Step 2: Extract Note Information ---
    print("\n--- Extracting Note Information ---")
    # Extract notes from the first instrument in the MIDI file [cite: 18]
    instrument = midi_data.instruments[0]
    notes = instrument.notes

    # Print the first 5 notes to see the structure
    print("First 5 notes (Pitch, Start Time, End Time, Velocity):")
    for i, note in enumerate(notes[:5]):
        print(f"Note {i+1}: Pitch={note.pitch}, Start={note.start:.2f}s, End={note.end:.2f}s, Velocity={note.velocity}")

    # --- Step 3: Convert to Piano-Roll Representation ---
    # The piano roll is a useful format for model input [cite: 19]
    fs = 10  # Sampling frequency for the piano roll (10 frames per second)
    piano_roll = midi_data.get_piano_roll(fs=fs)
    print(f"\nShape of the generated Piano-Roll: {piano_roll.shape} (Pitches, Time Steps)")

    # --- Step 4: Visualize the Piano Roll ---
    # Visualization helps in verifying the preprocessing steps [cite: 73]
    plt.figure(figsize=(12, 6))
    plt.imshow(piano_roll, aspect='auto', origin='lower', cmap='viridis')
    plt.title('Piano-Roll Representation of the MIDI File')
    plt.xlabel('Time Steps')
    plt.ylabel('MIDI Pitch')
    plt.colorbar(label='Velocity')
    plt.tight_layout()
    plt.show()
```

```python
# This example uses the 'mido' library for a lower-level view of MIDI events [cite: 42, 43]
import mido

def get_event_sequence(midi_file_path):
    """
    Parses a MIDI file and converts it into a sequence of events.
    """
    try:
        mid = mido.MidiFile(midi_file_path)
        events = []
        for msg in mid:
            # We only care about MIDI messages, not meta messages
            if not msg.is_meta:
                # Add time shifts between events
                if msg.time > 0:
                    events.append(f"TIME_SHIFT:{msg.time:.4f}")

                # Add Note-On or Note-Off events
                if msg.type == 'note_on':
                    events.append(f"NOTE_ON:{msg.note}")
                elif msg.type == 'note_off':
                    events.append(f"NOTE_OFF:{msg.note}")
        return events
    except Exception as e:
        print(f"Could not process MIDI file: {e}")
        return None

# --- Generate and Print Event Sequence ---
print("\n--- Generating Event-Based Representation ---")
event_sequence = get_event_sequence('/content/ACDC.Thunderstruck K.mid')

if event_sequence:
    print("Generated event-based sequence (first 20 events):")
    for event in event_sequence[:20]:
        print(event)
```

```python
# --- Step 1: Install all necessary tools, including pyfluidsynth ---
!apt-get install -y fluidsynth > /dev/null 2>&1
!pip install pretty_midi pyfluidsynth > /dev/null 2>&1
!wget -q https://github.com/musescore/MuseScore/raw/master/share/sound/FluidR3Mono_GM.sf2

import os
import pretty_midi
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

# --- Step 2: Define the corrected processing function ---
def visualize_midi_spectrogram(midi_file_path):
    """
    Loads a MIDI file, converts it to audio, and displays its Mel spectrogram.
    """
    print(f"--- Processing: {midi_file_path} ---")

    if not os.path.exists(midi_file_path):
        print(f"❌ ERROR: File not found at '{midi_file_path}'")
        return

    base_name = os.path.splitext(os.path.basename(midi_file_path))[0]
    output_wav_path = f"{base_name}.wav"
    soundfont_path = 'FluidR3Mono_GM.sf2'

    try:
        # 1. Synthesize MIDI to WAV
        print(f"Synthesizing '{midi_file_path}' to '{output_wav_path}'...")
        midi_data = pretty_midi.PrettyMIDI(midi_file_path)

        # CORRECTED: Using the new .synthesize() method to address the warning
        audio_data = midi_data.synthesize(fs=44100, sf2_path=soundfont_path)

        sf.write(output_wav_path, audio_data, 44100)

        # 2. Load the newly created audio file
        print("Loading audio and generating spectrogram...")
        y, sr = librosa.load(output_wav_path)

        # 3. Extract and visualize the Mel Spectrogram
        mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

        plt.figure(figsize=(12, 5))
        librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Mel Spectrogram for {os.path.basename(midi_file_path)}')
        plt.tight_layout()
        plt.show()
        print("✅ Done.\n")

    except Exception as e:
        print(f"An error occurred: {e}")


# --- Step 3: Use the function with your MIDI file ---
# Make sure you have uploaded the file to your Colab environment
visualize_midi_spectrogram('/content/ACDC.Thunderstruck K.mid')
```

```python
import librosa
import librosa.display

# --- Step 1: Load Audio File ---
# librosa comes with example audio files, which is convenient.
# We'll use the 'nutcracker' example clip.
try:
    y, sr = librosa.load(librosa.example('nutcracker'))
    print(f"Audio file loaded. Sample rate: {sr} Hz, Duration: {len(y)/sr:.2f}s")
except Exception as e:
    print(f"Could not load audio file: {e}")
    y, sr = None, None

if y is not None:
    # --- Step 2: Extract Mel Spectrogram and MFCCs ---
    # These are two of the most common features for audio AI tasks [cite: 22]

    # Mel Spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    # Convert to decibels (dB) for better visualization
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # MFCCs (Mel-Frequency Cepstral Coefficients)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13) # Usually 13-40 MFCCs are extracted

    print(f"\nShape of Mel Spectrogram: {mel_spectrogram_db.shape}")
    print(f"Shape of MFCCs: {mfccs.shape}")

    # --- Step 3: Normalize the Data ---
    # Normalization ensures uniformity, which is crucial for model training [cite: 23]
    # Simple min-max normalization for the MFCCs
    mfccs_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))
    print("\nMFCC data has been normalized.")

    # --- Step 4: Visualize the Audio Features ---
    # Using matplotlib and librosa.display to plot the features [cite: 71, 82]
    plt.figure(figsize=(12, 8))

    # Plot Mel Spectrogram
    plt.subplot(2, 1, 1)
    librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')

    # Plot MFCCs
    plt.subplot(2, 1, 2)
    librosa.display.specshow(mfccs_normalized, sr=sr, x_axis='time')
    plt.colorbar()
    plt.title('Normalized MFCCs') # As per your document's example [cite: 79]

    plt.tight_layout()
    plt.show()
```
